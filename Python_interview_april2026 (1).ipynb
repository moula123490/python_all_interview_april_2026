```python
# finding a duplicate in list

def dup_list(lis):
    seen=set()#---empty set
    dup=[]#---initiall duplicate list
    for i in lis:
        if i in seen and i not in dup:
            dup.append(i)
        else:
            seen.add(i)
    return ("duplicate is--->",dup,"and without duplicate are--->",seen)
    

dup_list([1,1,1,2,2,3,3,4,5,6,7])       
```

```python
#iqr method to detect outlier

Q1 = df["Salary"].quantile(0.25)
Q3 = df["Salary"].quantile(0.75)

IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

outliers = df[(df["Salary"] < lower) | (df["Salary"] > upper)]

print(outliers)
```

```python
'''
I first detect outliers using IQR or boxplots. If the outlier is due to data entry error,
I remove it. If it is a valid extreme value, I prefer capping or transforming the data 
so I don’t lose useful information. The choice depends on business context.
'''
```

```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore') 

q1=df['Salary'].quantile(0.25)
q3=df['Salary'].quantile(0.75)

IQR=q3-q1
upper_limit=q3+(1.5*IQR)
lower_limit=q1-(1.5*IQR)
print('upper_limit is--->',upper_limit)
outliers=df[(df['Salary']<lower_limit)|(df['Salary']>upper_limit)]['Salary']

df['Salary']=df['Salary'].clip(upper=upper_limit)
df
```

```python
# Automation helps reduce repetitive manual work, improves consistency, saves time, and reduces errors in recurring reporting tasks.

#frist i am creating function to clean the dataset

def clean_dataset(df):
    # to fill null values
    df['Sales']=df['Sales'].fillna(0)
    # to remove duplicates
    df=df.drop_duplicates()
    # to strip the spaces in name column
    df['Customer']=df['Customer'].str.strip()
    # convert date
    df["Date"] = pd.to_datetime(df["Date"], errors="coerce")
    return df
```

```python
files=[r"C:\Users\moula\Downloads\Chennai.csv",r"C:\Users\moula\Downloads\banglore.csv",r"C:\Users\moula\Downloads\Hyderabad.csv"]
add_csv=[]
for file in files:
    #read csv
    df=pd.read_csv(file)
    #clean csv
    df=clean_dataset(df)
    #appendcsv
    add_csv.append(df)
final_dataset=pd.concat(add_csv)
final_dataset
```
